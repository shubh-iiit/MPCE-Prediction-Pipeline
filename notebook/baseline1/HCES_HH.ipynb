```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load household-level data
train_HH_df = pd.read_parquet("/content/train/HH_Level_train_data.parquet")

# Convert binary columns to numeric 0/1
binary_cols = [col for col in train_HH_df.columns if col.startswith("Is_")]
train_HH_df[binary_cols] = train_HH_df[binary_cols].apply(pd.to_numeric, errors='coerce').fillna(0)

# Convert HH Size and other numerics
train_HH_df["HH Size (For FDQ)"] = pd.to_numeric(train_HH_df["HH Size (For FDQ)"], errors='coerce')
train_HH_df["TotalExpense"] = pd.to_numeric(train_HH_df["TotalExpense"], errors='coerce')

# ========== 1. Sector Distribution ==========
plt.figure(figsize=(6, 4))
ax = sns.countplot(x="Sector", data=train_HH_df, palette="Set2")
plt.title("Sector Distribution (1=Rural, 2=Urban)")
for p in ax.patches:
    ax.annotate(f'{int(p.get_height()):,}', (p.get_x() + 0.2, p.get_height() + 1000))
plt.tight_layout()
plt.show()

# ========== 2. State Distribution (top 10) ==========
top_states = train_HH_df["State"].value_counts().nlargest(10).index
plt.figure(figsize=(10, 5))
ax = sns.countplot(x="State", data=train_HH_df[train_HH_df["State"].isin(top_states)], order=top_states, palette="Paired")
plt.title("Top 10 States by Household Count")
for p in ax.patches:
    ax.annotate(f'{int(p.get_height()):,}', (p.get_x() + 0.2, p.get_height() + 1000))
plt.tight_layout()
plt.show()

# ========== 3. HH Size Distribution ==========
plt.figure(figsize=(8, 5))
sns.histplot(train_HH_df["HH Size (For FDQ)"].dropna(), bins=range(1, 20), color='coral', kde=False)
plt.title("Household Size Distribution")
plt.xlabel("Household Size")
plt.ylabel("Count")
plt.tight_layout()
plt.show()

# ========== 4. Total Expense Distribution ==========
plt.figure(figsize=(10, 5))
sns.histplot(train_HH_df["TotalExpense"].dropna(), bins=50, color='skyblue', kde=True)
plt.title("Total Household Expense Distribution")
plt.xlabel("Total Expense")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()

# ========== 5. Average Expense by Social Group (top 5 groups) ==========
top_social_groups = train_HH_df["Social Group of the head of the household"].value_counts().nlargest(5).index
sg_exp = train_HH_df[train_HH_df["Social Group of the head of the household"].isin(top_social_groups)]
sg_avg_exp = sg_exp.groupby("Social Group of the head of the household")["TotalExpense"].mean().sort_values()

plt.figure(figsize=(8, 4))
ax = sns.barplot(x=sg_avg_exp.index.astype(str), y=sg_avg_exp.values, palette="Blues_d")
plt.title("Average Total Expense by Top 5 Social Groups")
plt.xlabel("Social Group")
plt.ylabel("Average Total Expense")
for i, v in enumerate(sg_avg_exp.values):
    ax.text(i, v + 500, f"{v:,.0f}", ha='center')
plt.tight_layout()
plt.show()

# ========== 6. Ownership of Key Assets (binary) ==========
assets = [
    "Is_HH_Have_Television",
    "Is_HH_Have_Radio",
    "Is_HH_Have_Laptop_PC",
    "Is_HH_Have_Mobile_handset",
    "Is_HH_Have_Refrigerator",
    "Is_HH_Have_Washing_machine",
    "Is_HH_Have_Airconditioner_aircooler"
]

ownership = train_HH_df[assets].mean().sort_values(ascending=False) * 100  # percent ownership

plt.figure(figsize=(10, 5))
ax = sns.barplot(x=ownership.index.str.replace("Is_HH_Have_", "").str.replace("_", " "),
                 y=ownership.values, palette="muted")
plt.title("Ownership Percentage of Key Household Assets")
plt.ylabel("Ownership (%)")
plt.xticks(rotation=45)
for i, v in enumerate(ownership.values):
    ax.text(i, v + 1, f"{v:.1f}%", ha='center')
plt.tight_layout()
plt.show()

# ========== 7. Correlation Heatmap of Numeric + Binary ==========
numeric_cols = ["HH Size (For FDQ)", "TotalExpense"] + assets
numeric_df = train_HH_df[numeric_cols].copy()
numeric_df = numeric_df.apply(pd.to_numeric, errors='coerce')

plt.figure(figsize=(10, 8))
corr = numeric_df.corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f", square=True)
plt.title("Correlation Heatmap of Household Size, Expense, and Asset Ownership")
plt.tight_layout()
plt.show()
```

# Step 2: Combine with Aggregated Person-Level Data
First, aggregate person-level data per household (HH_ID):

```python
# Load person-level data
person_df = pd.read_parquet("/content/train/Person_Level_Train_Data.parquet")

# Convert numeric columns for aggregation
person_df["Age(in years)"] = pd.to_numeric(person_df["Age(in years)"], errors='coerce')
person_df["Total year of education completed"] = pd.to_numeric(person_df["Total year of education completed"], errors='coerce')
person_df["Whether used internet from any location during last 30 days"] = pd.to_numeric(
    person_df["Whether used internet from any location during last 30 days"], errors='coerce'
).fillna(0)

# Aggregate per household
person_agg = person_df.groupby("HH_ID").agg(
    avg_age=("Age(in years)", "mean"),
    avg_education=("Total year of education completed", "mean"),
    internet_users=("Whether used internet from any location during last 30 days", lambda x: (x == 1).sum()),
    total_members=("Person Srl No.", "count")
).reset_index()

print("Aggregated person-level data per household:")
print(person_agg.head())
```

```python
# Merge with household-level data
hh_person_merged = train_HH_df.merge(person_agg, on="HH_ID", how="left")

# Quick check of merged data
print("Merged Household + Person data shape:", hh_person_merged.shape)
print(hh_person_merged.head())
```

```python
# Plot average age vs. TotalExpense by Sector
plt.figure(figsize=(8, 5))
sns.scatterplot(data=hh_person_merged, x="avg_age", y="TotalExpense", hue="Sector", palette="Set1", alpha=0.6)
plt.title("Household Average Age vs Total Expense by Sector")
plt.xlabel("Average Age of Household Members")
plt.ylabel("Total Expense")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1) Recompute pivot tables (if not already in memory)
train = pd.read_parquet("/content/train/HH_Level_train_data.parquet")
train["Sector"] = pd.to_numeric(train["Sector"], errors="coerce")
pivot_state = train.pivot_table(
    index="State", columns="Sector",
    values="TotalExpense", aggfunc="mean"
).rename(columns={1: "Rural", 2: "Urban"}).reset_index()
pivot_district = train.pivot_table(
    index="District", columns="Sector",
    values="TotalExpense", aggfunc="mean"
).rename(columns={1: "Rural", 2: "Urban"}).reset_index()

# Melt them for easy plotting
state_melt = pivot_state.melt(id_vars="State",
                              value_vars=["Rural", "Urban"],
                              var_name="Sector",
                              value_name="AvgExpense")
district_melt = pivot_district.melt(id_vars="District",
                                    value_vars=["Rural", "Urban"],
                                    var_name="Sector",
                                    value_name="AvgExpense")

sns.set(style="whitegrid")

# 2) Comparative bar chart: State (all)
plt.figure(figsize=(14, 5))
ax = sns.barplot(data=state_melt, x="State", y="AvgExpense", hue="Sector", palette="Set1")
ax.set_title("Actual Average TotalExpense by State: Urban vs Rural")
ax.set_xlabel("State")
ax.set_ylabel("Avg TotalExpense")
plt.xticks(rotation=45)
plt.legend(title="Sector")
plt.tight_layout()
plt.show()

# 3) Comparative line plot: State
plt.figure(figsize=(14, 4))
sns.lineplot(data=pivot_state, x="State", y="Urban", marker="o", label="Urban")
sns.lineplot(data=pivot_state, x="State", y="Rural", marker="o", label="Rural")
plt.title("Actual Urban vs Rural Avg Expense by State")
plt.xlabel("State")
plt.ylabel("Avg TotalExpense")
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()

# 4) Comparative bar chart: District (all) — might be very wide
plt.figure(figsize=(16, 6))
ax = sns.barplot(data=district_melt, x="District", y="AvgExpense", hue="Sector", palette="Set2")
ax.set_title("Actual Average TotalExpense by District: Urban vs Rural")
ax.set_xlabel("District")
ax.set_ylabel("Avg TotalExpense")
plt.xticks(rotation=90)
plt.legend(title="Sector")
plt.tight_layout()
plt.show()

# 5) Comparative line plot: District
plt.figure(figsize=(14, 4))
sns.lineplot(data=pivot_district, x="District", y="Urban", marker="o", label="Urban")
sns.lineplot(data=pivot_district, x="District", y="Rural", marker="o", label="Rural")
plt.title("Actual Urban vs Rural Avg Expense by State")
plt.xlabel("District")
plt.ylabel("Avg TotalExpense")
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()

# 5) Scatter: Urban vs Rural by District
plt.figure(figsize=(6, 6))
plt.scatter(pivot_district["Urban"], pivot_district["Rural"], alpha=0.6)
plt.plot([pivot_district[["Urban","Rural"]].min().min(),
          pivot_district[["Urban","Rural"]].max().max()],
         [pivot_district[["Urban","Rural"]].min().min(),
          pivot_district[["Urban","Rural"]].max().max()],
         '--r')
plt.title("Actual Urban vs Rural Avg Expense by District")
plt.xlabel("Urban AvgExpense")
plt.ylabel("Rural AvgExpense")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

# Load data
train_HH_df = pd.read_parquet("/content/train/HH_Level_train_data.parquet")

# Identify binary columns
binary_cols = [col for col in train_HH_df.columns if col.startswith("Is_")]
train_HH_df[binary_cols] = train_HH_df[binary_cols].apply(pd.to_numeric, errors='coerce')

# Separate features and ID
df = train_HH_df.drop(columns=["HH_ID"])  # Drop ID for preprocessing

# --- 1. Split columns ---
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns.tolist()
categorical_cols = df.select_dtypes(include='object').columns.tolist()

# Remove binary columns from categorical (already numeric)
categorical_cols = [col for col in categorical_cols if col not in binary_cols]

# --- 2. Pipelines ---

# Pipeline for numeric features
numeric_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

# Pipeline for categorical features
categorical_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', OneHotEncoder(handle_unknown='ignore', sparse_output=False)) # Changed sparse to sparse_output
])

# Combine using ColumnTransformer
preprocessor = ColumnTransformer(transformers=[
    ('num', numeric_pipeline, numeric_cols),
    ('cat', categorical_pipeline, categorical_cols)
])

# --- 3. Fit and Transform ---
processed_array = preprocessor.fit_transform(df)

# Get feature names
encoded_cat_cols = preprocessor.named_transformers_['cat']['encoder'].get_feature_names_out(categorical_cols)
final_feature_names = numeric_cols + encoded_cat_cols.tolist()

# Final DataFrame
processed_df = pd.DataFrame(processed_array, columns=final_feature_names)

# Optional: Check result
print("Processed DataFrame shape:", processed_df.shape)
print(processed_df.head())
```

# Feature Engineering

* Enriched household data with person-level aggregate context

* Frequency encoding for high-cardinality categorical features

* One-hot encoding for low-cardinality features

* Proper scaling/imputation for numeric features

```python
import pandas as pd
import numpy as np
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.preprocessing import FunctionTransformer, OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

# ----------------------------
# 1. Define helper functions
# ----------------------------

def load_and_merge(hh_path, person_path):
    hh_df = pd.read_parquet(hh_path)
    person_df = pd.read_parquet(person_path)

    # Convert binary flags
    binary_cols = [col for col in hh_df.columns if col.startswith("Is_")]
    hh_df[binary_cols] = hh_df[binary_cols].apply(pd.to_numeric, errors='coerce')

    # Aggregate person-level data
    person_df["Age(in years)"] = pd.to_numeric(person_df["Age(in years)"], errors='coerce')
    person_df["Total year of education completed"] = pd.to_numeric(person_df["Total year of education completed"], errors='coerce')
    person_df["Whether used internet from any location during last 30 days"] = pd.to_numeric(
        person_df["Whether used internet from any location during last 30 days"], errors='coerce'
    ).fillna(0)

    person_agg = person_df.groupby("HH_ID").agg(
        avg_age=("Age(in years)", "mean"),
        avg_education=("Total year of education completed", "mean"),
        internet_users=("Whether used internet from any location during last 30 days", lambda x: (x == 1).sum()),
        total_members=("Person Srl No.", "count")
    ).reset_index()

    # Merge
    merged = hh_df.merge(person_agg, on="HH_ID", how="left")
    return merged

# ----------------------------
# 2. Load train and test merged datasets
# ----------------------------

train = load_and_merge("/content/train/HH_Level_train_data.parquet",
                       "/content/train/Person_Level_Train_Data.parquet")

test = load_and_merge("/content/test/HH_Level_test_data.parquet",
                      "/content/test/Person_Level_Test_Data.parquet")

# ----------------------------
# 3. Set up feature engineering
# ----------------------------

def add_nic2d(df):
    df = df.copy()
    df['NIC_2D'] = df['NIC_5D'].astype(str).str[:2]
    return df.drop(columns=['NIC_5D'])

class FrequencyEncoder(BaseEstimator, TransformerMixin):
    def __init__(self, columns):
        self.columns = columns
    def fit(self, X, y=None):
        self.freq_maps_ = {col: X[col].value_counts(normalize=True).to_dict() for col in self.columns}
        return self
    def transform(self, X):
        X = X.copy()
        for col, fmap in self.freq_maps_.items():
            X[col] = X[col].map(fmap).fillna(0)
        return X[self.columns]

# ----------------------------
# 4. Prepare inputs for pipeline
# ----------------------------

binary_cols = [col for col in train.columns if col.startswith("Is_")]
person_feats = ['avg_age', 'avg_education', 'internet_users', 'total_members']
numeric_cols = binary_cols + person_feats
categorical_cols = train.select_dtypes(include='object').columns.tolist()

high_card = ['NIC_2D', 'District']
low_card = [c for c in categorical_cols if c not in high_card + ['NIC_5D', 'HH_ID']]

# ----------------------------
# 5. Define pipeline
# ----------------------------

numeric_pipeline = Pipeline([
    ('impute', SimpleImputer(strategy='median')),
    ('scale', StandardScaler())
])

high_card_pipeline = Pipeline([
    ('freq', FrequencyEncoder(high_card))
])

low_card_pipeline = Pipeline([
    ('impute', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])

preprocessor = ColumnTransformer([
    ('num', numeric_pipeline, numeric_cols),
    ('high', high_card_pipeline, high_card),
    ('low',  low_card_pipeline,  low_card)
])

full_pipeline = Pipeline([
    ('nic2d', FunctionTransformer(add_nic2d, validate=False)),
    ('preproc', preprocessor),
])

# ----------------------------
# 6. Fit on train, transform both
# ----------------------------

X_train_full = train.drop(columns=["HH_ID", "TotalExpense"])
y_train = train["TotalExpense"]

X_test_full = test.drop(columns=["HH_ID"])

X_train = full_pipeline.fit_transform(X_train_full)
X_test = full_pipeline.transform(X_test_full)

print("✅ Final shapes:")
print("X_train:", X_train.shape)
print("X_test:", X_test.shape)
```

# Training pipeline
* Log-transform the target: log(TotalExpense + 1)

* Fit a linear regression model

* Predict on test set, then inverse-transform predictions: exp(y_pred) - 1

* Evaluate using RMSE and R² in original scale

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_percentage_error
import numpy as np
import pandas as pd

# 1. Log-transform the training target
y_train_log = np.log1p(y_train)

# 2. Train linear regression model on preprocessed full training set
lr = LinearRegression()
lr.fit(X_train, y_train_log)

# 3. Predict in log space on test set
y_pred_log = lr.predict(X_test)

# 4. Inverse-transform predictions to original scale
y_pred = np.expm1(y_pred_log)

# (Optional) If you have ground truth for the test set:
# y_test = test["TotalExpense"]  # If available
# y_test = np.array(y_test)

# 5. Create results DataFrame (if HH_IDs in test are available)
results_df = pd.DataFrame({
    "HH_ID": test["HH_ID"].values,
    "Predicted_TotalExpense": y_pred
})

# 6. Show summary
print(f"\n Model trained on full training data.")
print(f" Predictions made on test set of shape: {X_test.shape}")
```

## Notes:

* Linear regression assumes additive effects in log space, which is equivalent to multiplicative effects in real space — suitable for skewed distributions.

* This model is very interpretable and forms a good baseline.
