```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set global style
sns.set(style="whitegrid", palette="pastel")

# Mappings
gender_map = {1: "Male", 2: "Female", 3: "Other"}
marital_map = {
    1: "Never Married",
    2: "Married",
    3: "Widowed",
    4: "Divorced/Separated"
}
relation_map = {
    1: "Self", 2: "Spouse", 3: "Married child", 4: "Spouse of married child",
    5: "Unmarried child", 6: "Grandchild", 7: "Parent/In-law",
    8: "Sibling/In-law", 9: "Others"
}

# Clean up
df = pd.read_parquet("/content/train/Person_Level_Train_Data.parquet")
df["Gender"] = df["Gender"].astype(int).map(gender_map)
df["Marital Status"] = df["Marital Status (code)"].astype(int).map(marital_map)
df["Relation to Head"] = df["Relation to head (code)"].astype(int).map(relation_map)

# Combine rare categories in Relation to Head
rel_counts = df["Relation to Head"].value_counts()
major_rels = rel_counts[rel_counts > 20000].index
df["Relation to Head"] = df["Relation to Head"].apply(lambda x: x if x in major_rels else "Others")

# ========== 🟦 GENDER DISTRIBUTION ==========
plt.figure(figsize=(6, 4))
ax = sns.countplot(x="Gender", data=df)
plt.title("Gender Distribution")
for p in ax.patches:
    ax.annotate(f'{int(p.get_height()):,}', (p.get_x() + 0.3, p.get_height() + 1000))
plt.tight_layout()
plt.show()

# ========== MARITAL STATUS ==========
plt.figure(figsize=(7, 4))
order = ["Never Married", "Married", "Widowed", "Divorced/Separated"]
ax = sns.countplot(x="Marital Status", data=df, order=order)
plt.title("Marital Status Distribution")
for p in ax.patches:
    ax.annotate(f'{int(p.get_height()):,}', (p.get_x() + 0.15, p.get_height() + 1000))
plt.xticks(rotation=10)
plt.tight_layout()
plt.show()

# ========== RELATION TO HEAD ==========
plt.figure(figsize=(10, 5))
ax = sns.countplot(x="Relation to Head", data=df, order=df["Relation to Head"].value_counts().index)
plt.title("Relation to Head of Household")
for p in ax.patches:
    ax.annotate(f'{int(p.get_height()):,}', (p.get_x() + 0.1, p.get_height() + 1000))
plt.xticks(rotation=15)
plt.tight_layout()
plt.show()

# ========== AGE DISTRIBUTION ==========
df["Age(in years)"] = pd.to_numeric(df["Age(in years)"], errors="coerce")
plt.figure(figsize=(10, 5))
sns.histplot(df["Age(in years)"], bins=40, kde=True, color='skyblue')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.grid(True)
plt.tight_layout()
plt.show()

# ========== EDUCATION YEARS ==========
df["Total year of education completed"] = pd.to_numeric(df["Total year of education completed"], errors="coerce")
plt.figure(figsize=(10, 5))
sns.histplot(df["Total year of education completed"], bins=25, kde=True, color='mediumpurple')
plt.title("Years of Education Completed")
plt.xlabel("Years")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_parquet("/content/train/Person_Level_Train_Data.parquet")

# 1. Convert key columns to numeric
df["Age(in years)"] = pd.to_numeric(df["Age(in years)"], errors="coerce")
df["Total year of education completed"] = pd.to_numeric(df["Total year of education completed"], errors="coerce")
df["No. of days stayed away from home during last 30 days"] = pd.to_numeric(
    df["No. of days stayed away from home during last 30 days"], errors="coerce"
)
df["Whether used internet from any location during last 30 days"] = pd.to_numeric(
    df["Whether used internet from any location during last 30 days"], errors="coerce"
).fillna(0)

# Meal columns to numeric
meal_cols = [
    "No. of meals usually taken in a day",
    "No. of meals taken during last 30 days from school, balwadi etc.",
    "No. of meals taken during last 30 days from employer as perquisites or part of wage",
    "No. of meals taken during last 30 days  others",
    "No. of meals taken during last 30 days on payment",
    "No. of meals taken during last 30 days at home",
]
df[meal_cols] = df[meal_cols].apply(pd.to_numeric, errors="coerce")

# 2. Mapping code values
gender_map = {"1": "Male", "2": "Female", "3": "Other", 1: "Male", 2: "Female", 3: "Other"}
marital_map = {
    "1": "Never married",
    "2": "Currently married",
    "3": "Widowed",
    "4": "Divorced/Separated",
    1: "Never married",
    2: "Currently married",
    3: "Widowed",
    4: "Divorced/Separated",
}
relation_map = {
    "1": "Self",
    "2": "Spouse",
    "3": "Married child",
    "4": "Spouse of married child",
    "5": "Unmarried child",
    "6": "Grandchild",
    "7": "Father/Mother/Father-in-law/Mother-in-law",
    "8": "Brother/Sister/Bro-in-law/Sis-in-law",
    "9": "Others",
    1: "Self",
    2: "Spouse",
    3: "Married child",
    4: "Spouse of married child",
    5: "Unmarried child",
    6: "Grandchild",
    7: "Parents/In-laws",
    8: "Sibling/In-laws",
    9: "Others"
}

df["Gender"] = df["Gender"].map(gender_map)
df["Marital Status"] = df["Marital Status (code)"].map(marital_map)
df["Relation to Head"] = df["Relation to head (code)"].map(relation_map)

# Box: Meals on Payment vs Marital Status
sns.boxplot(data=df, x="Marital Status", y="No. of meals taken during last 30 days on payment")
plt.title("Meals Taken on Payment vs Marital Status")
plt.xticks(rotation=20)
plt.show()

# Internet usage by gender (cleaned)
plt.figure(figsize=(8, 6))
internet_gender = pd.crosstab(df["Gender"], df["Whether used internet from any location during last 30 days"])
internet_gender.plot(kind="bar", stacked=True)
plt.title("Internet Usage by Gender")
plt.ylabel("Count")
plt.xlabel("Gender")
plt.legend(title="Internet Usage (1=yes, 2=no, 0=missing)")
plt.show()

# Age vs Internet Usage (Scatter/Violin)
sns.violinplot(x="Whether used internet from any location during last 30 days", y="Age(in years)", data=df)
plt.title("Age vs Internet Usage")
plt.xlabel("Internet Usage")
plt.ylabel("Age")
plt.show()

# 4. Aggregate by HH_ID
agg_df = df.groupby("HH_ID").agg(
    avg_age=("Age(in years)", "mean"),
    avg_education=("Total year of education completed", "mean"),
    internet_users=("Whether used internet from any location during last 30 days", lambda x: (x == 1).sum()),
    total_members=("Person Srl No.", "count"),
    avg_meals_at_home=("No. of meals taken during last 30 days at home", "mean")
).reset_index()

print("Aggregated Household Summary:")
print(agg_df.head())
```

## Merge & Run Person-Level EDA with Household Metadata
* Clean merge between HH and Person levels.
* Aggregated summaries:
* By Social Group: Age, education

* By Religion: Education, internet usage rate

* By NSS Region: Age, internet access

* Full correlation heatmap across person-level numeric features and HH metadata.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load both datasets
hh_df = pd.read_parquet("/content/train/HH_Level_train_data.parquet")
person_df = pd.read_parquet("/content/train/Person_Level_Train_Data.parquet")

# Select household metadata to merge
hh_meta = hh_df[[
    "HH_ID",
    "Social Group of the head of the household",
    "Religion of the head of the household",
    "NSS-Region"
]]

# Merge person-level with household-level metadata
df = person_df.merge(hh_meta, on="HH_ID", how="left")

# Convert numeric columns
df["Age(in years)"] = pd.to_numeric(df["Age(in years)"], errors='coerce')
df["Total year of education completed"] = pd.to_numeric(df["Total year of education completed"], errors='coerce')
df["Whether used internet from any location during last 30 days"] = pd.to_numeric(
    df["Whether used internet from any location during last 30 days"], errors='coerce').fillna(0)
df["No. of days stayed away from home during last 30 days"] = pd.to_numeric(
    df["No. of days stayed away from home during last 30 days"], errors='coerce')

# Clean meal columns
meal_cols = [
    "No. of meals usually taken in a day",
    "No. of meals taken during last 30 days from school, balwadi etc.",
    "No. of meals taken during last 30 days from employer as perquisites or part of wage",
    "No. of meals taken during last 30 days  others",
    "No. of meals taken during last 30 days on payment",
    "No. of meals taken during last 30 days at home"
]
df[meal_cols] = df[meal_cols].apply(pd.to_numeric, errors='coerce')

# === 1. Summary by Social Group ===
sg_summary = df.groupby("Social Group of the head of the household").agg(
    AvgAge=("Age(in years)", "mean"),
    AvgEducation=("Total year of education completed", "mean"),
    Count=("HH_ID", "count")
).reset_index()
print("\nAverage Age & Education by Social Group")
print(sg_summary)

# === 2. Summary by Religion ===
religion_summary = df.groupby("Religion of the head of the household").agg(
    AvgEducation=("Total year of education completed", "mean"),
    InternetUsageRate=("Whether used internet from any location during last 30 days", lambda x: (x == 1).mean()),
    Count=("HH_ID", "count")
).reset_index()
print("\nEducation & Internet Usage by Religion")
print(religion_summary)

# === 3. Summary by NSS Region ===
region_summary = df.groupby("NSS-Region").agg(
    InternetUsageRate=("Whether used internet from any location during last 30 days", lambda x: (x == 1).mean()),
    AvgAge=("Age(in years)", "mean"),
    Count=("HH_ID", "count")
).reset_index()
print("\nInternet Usage & Age by NSS Region")
print(region_summary)

# === 4. Correlation Heatmap ===
numeric_df = df.select_dtypes(include=[np.number])
corr = numeric_df.corr()

plt.figure(figsize=(12, 10))
sns.heatmap(corr, cmap="coolwarm", annot=True, fmt=".2f", square=True)
plt.title("Feature Correlation Heatmap (Person-Level with HH Metadata)")
plt.tight_layout()
plt.show()
```
