# Top-k features Used for Classsification(by importance)
* prepare_features_batch is your fast, oneâ€shot encoder+scaler for the entire DataFrame (youâ€™ll need to add this helper if you donâ€™t have it; it was shown earlier).

* We drop both Income_Class (the target) and TotalExpense (the leakage) before feature prep.

* The RandomForestâ€™s feature_importances_ reflect the relative Giniâ€importance of each column in splitting the 4 classes.

* Can swap in an XGBClassifier the same wayâ€”just call its .feature_importances_ after fitting.

```python
# feature_importance_classifiers.py

import pandas as pd
import numpy as np
import joblib

# â”€â”€â”€ HELPERS â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
def get_feature_names(feature_info):
    names = []
    for col in feature_info["categorical_cols"]:
        cats = feature_info["encoders"][col].categories_[0]
        names += [f"{col}_{cat}" for cat in cats]
    names += feature_info["numerical_cols"]
    return names

def print_topk_importances(importances, feature_names, model_name, k=20):
    indices = np.argsort(importances)[::-1][:k]
    print(f"\nTop {k} features for {model_name}:\n")
    for rank, idx in enumerate(indices, 1):
        print(f"{rank:2d}. {feature_names[idx]:<40} {importances[idx]:.4f}")

# â”€â”€â”€ CONFIG â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
CLASSIFIER_PICKLE = "/content/drive/MyDrive/MPCE_MoSPI/Shubham-IIIT/models_Clf/sector_income_classifiers_tuned.pkl"
REGRESSOR_PICKLE  = "/content/drive/MyDrive/MPCE_MoSPI/Shubham-IIIT/models_Regressor/sector_income_randomforestmodel.pkl"  # to grab feature_info
TOP_K             = 20

# â”€â”€â”€ 1. Load feature_info from your regressor pickle â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
reg_data    = joblib.load(REGRESSOR_PICKLE)
feature_info= reg_data["feature_info"]
feature_names = get_feature_names(feature_info)

# â”€â”€â”€ 2. Load the tuned classifiers â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
clf_data  = joblib.load(CLASSIFIER_PICKLE)
clf_rural = clf_data["clf_rural"]
clf_urban = clf_data["clf_urban"]

# â”€â”€â”€ 3. Extract and print importances â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
print_topk_importances(clf_rural.feature_importances_, feature_names, "Rural Classifier", TOP_K)
print_topk_importances(clf_urban.feature_importances_, feature_names, "Urban Classifier", TOP_K)
```

# Difference in Top-K features in Rural/Urban Classifier

```python
def get_topk_features_list(importances, feature_names, k=20):
    indices = np.argsort(importances)[::-1][:k]
    return [feature_names[idx] for idx in indices]

# â”€â”€â”€ 4. Get top K features as lists â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
top_rural_features = get_topk_features_list(clf_rural.feature_importances_, feature_names, TOP_K)
top_urban_features = get_topk_features_list(clf_urban.feature_importances_, feature_names, TOP_K)

# â”€â”€â”€ 5. Convert lists to sets â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
set_rural = set(top_rural_features)
set_urban = set(top_urban_features)

# â”€â”€â”€ 6. Find the differences â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
# Features in Rural set but not in Urban set
diff_rural_not_urban = set_rural - set_urban

# Features in Urban set but not in Rural set
diff_urban_not_rural = set_urban - set_rural

# â”€â”€â”€ 7. Print the differences â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
print(f"\nFeatures in Rural top {TOP_K} but NOT in Urban top {TOP_K}:")
for feature in diff_rural_not_urban:
    print(f"- {feature}")

print(f"\nFeatures in Urban top {TOP_K} but NOT in Rural top {TOP_K}:")
for feature in diff_urban_not_rural:
    print(f"- {feature}")
```

# Correlation of Top-K features with Income Classes{1: Lower Rural, 2: Upper Rural, 3: Lower Urban, 4: Upper Urban}

```python
import pandas as pd
import numpy as np
import joblib

# â”€â”€â”€ HELPERS â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
def get_feature_names(feature_info):
    names = []
    for col in feature_info["categorical_cols"]:
        cats = feature_info["encoders"][col].categories_[0]
        names += [f"{col}_{cat}" for cat in cats]
    names += feature_info["numerical_cols"]
    return names

def topk_features(importances, feature_names, k=20):
    idxs = np.argsort(importances)[::-1][:k]
    return [feature_names[i] for i in idxs]

# â”€â”€â”€ CONFIG â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
CLASSIFIER_PICKLE = "/content/drive/MyDrive/MPCE_MoSPI/Shubham-IIIT/models_Clf/sector_income_classifiers_tuned.pkl"
REGRESSOR_PICKLE  = "/content/drive/MyDrive/MPCE_MoSPI/Shubham-IIIT/models_Regressor/sector_income_randomforestmodel.pkl"
TRAIN_WITH_CLASS  = "/content/drive/MyDrive/MPCE_MoSPI/Shubham-IIIT/train_with_class.csv"
TOP_K             = 20

# â”€â”€â”€ 1. Load feature_info and feature names â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
reg_data      = joblib.load(REGRESSOR_PICKLE)
feature_info  = reg_data["feature_info"]
feature_names = get_feature_names(feature_info)

# â”€â”€â”€ 2. Load classifiers â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
clf_data  = joblib.load(CLASSIFIER_PICKLE)
clf_rural = clf_data["clf_rural"]
clf_urban = clf_data["clf_urban"]

# â”€â”€â”€ 3. Identify top K features per classifier â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
top_rural = topk_features(clf_rural.feature_importances_, feature_names, TOP_K)
top_urban = topk_features(clf_urban.feature_importances_, feature_names, TOP_K)

# â”€â”€â”€ 4. Load full training set â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
df = pd.read_csv(TRAIN_WITH_CLASS)

# â”€â”€â”€ 5. Compute correlations for Rural (classes 1 vs 2) â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
rural_df = df[df["Income_Class"].isin([1, 2])].dropna(subset=top_rural)
# Binary target: 1 = upper (class 2), 0 = lower (class 1)
rural_df["is_upper"] = (rural_df["Income_Class"] == 2).astype(int)

rural_corrs = []
for feat in top_rural:
    corr = rural_df[[feat, "is_upper"]].corr().loc[feat, "is_upper"]
    rural_corrs.append((feat, corr))
rural_corrs = sorted(rural_corrs, key=lambda x: abs(x[1]), reverse=True)

print(f"\nRural Classifier Top {TOP_K} Feature Correlation with (Upper Rural=1):\n")
for feat, corr in rural_corrs:
    print(f"{feat:<40s} {corr:.4f}")

# â”€â”€â”€ 6. Compute correlations for Urban (classes 3 vs 4) â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
urban_df = df[df["Income_Class"].isin([3, 4])].dropna(subset=top_urban)
# Binary target: 1 = upper (class 4), 0 = lower (class 3)
urban_df["is_upper"] = (urban_df["Income_Class"] == 4).astype(int)

urban_corrs = []
for feat in top_urban:
    corr = urban_df[[feat, "is_upper"]].corr().loc[feat, "is_upper"]
    urban_corrs.append((feat, corr))
urban_corrs = sorted(urban_corrs, key=lambda x: abs(x[1]), reverse=True)

print(f"\nUrban Classifier Top {TOP_K} Feature Correlation with (Upper Urban=1):\n")
for feat, corr in urban_corrs:
    print(f"{feat:<40s} {corr:.4f}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import joblib

# â”€â”€â”€ LOAD IMPORTS & HELPERS â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
def get_feature_names(feature_info):
    names = []
    for col in feature_info["categorical_cols"]:
        cats = feature_info["encoders"][col].categories_[0]
        names += [f"{col}_{cat}" for cat in cats]
    names += feature_info["numerical_cols"]
    return names

def topk_features(importances, feature_names, k=20):
    idxs = np.argsort(importances)[::-1][:k]
    return [feature_names[i] for i in idxs]

# â”€â”€â”€ CONFIG â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
CLASSIFIER_PICKLE = "/content/drive/MyDrive/MPCE_MoSPI/Shubham-IIIT/models_Clf/sector_income_classifiers_tuned.pkl"
REGRESSOR_PICKLE  = "/content/drive/MyDrive/MPCE_MoSPI/Shubham-IIIT/models_Regressor/sector_income_randomforestmodel.pkl"
TRAIN_WITH_CLASS  = "/content/drive/MyDrive/MPCE_MoSPI/Shubham-IIIT/train_with_class.csv"
TOP_K             = 20

# â”€â”€â”€ 1. Load feature_info & feature names â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
reg = joblib.load(REGRESSOR_PICKLE)
feature_info  = reg["feature_info"]
feature_names = get_feature_names(feature_info)

# â”€â”€â”€ 2. Load classifiers & pick top features â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
clf = joblib.load(CLASSIFIER_PICKLE)
clf_rural = clf["clf_rural"]
clf_urban = clf["clf_urban"]

top_rural = topk_features(clf_rural.feature_importances_, feature_names, TOP_K)
top_urban = topk_features(clf_urban.feature_importances_, feature_names, TOP_K)

# â”€â”€â”€ 3. Load data and compute correlations â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
df = pd.read_csv(TRAIN_WITH_CLASS)

# Rural correlations
r = df[df["Income_Class"].isin([1,2])].dropna(subset=top_rural).copy()
r["is_upper"] = (r["Income_Class"]==2).astype(int)
r_corrs = {feat: r[[feat,"is_upper"]].corr().loc[feat,"is_upper"] for feat in top_rural}
r_corrs = pd.Series(r_corrs).abs().sort_values(ascending=True)  # sort for barh

# Urban correlations
u = df[df["Income_Class"].isin([3,4])].dropna(subset=top_urban).copy()
u["is_upper"] = (u["Income_Class"]==4).astype(int)
u_corrs = {feat: u[[feat,"is_upper"]].corr().loc[feat,"is_upper"] for feat in top_urban}
u_corrs = pd.Series(u_corrs).abs().sort_values(ascending=True)

# â”€â”€â”€ 4. Plot â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
fig, axes = plt.subplots(1,2, figsize=(12,8), sharex=False)

# Rural plot
axes[0].barh(r_corrs.index, r_corrs.values, color="#4c72b0")
axes[0].set_title("Rural: |corr(feature, is_upper)|")
axes[0].set_xlabel("Absolute Pearson r")
axes[0].tick_params(axis='y', labelsize=8)

# Urban plot
axes[1].barh(u_corrs.index, u_corrs.values, color="#c44e52")
axes[1].set_title("Urban: |corr(feature, is_upper)|")
axes[1].set_xlabel("Absolute Pearson r")
axes[1].tick_params(axis='y', labelsize=8)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import joblib

# â”€â”€â”€ HELPERS â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
def get_feature_names(feature_info):
    names = []
    for col in feature_info["categorical_cols"]:
        cats = feature_info["encoders"][col].categories_[0]
        names += [f"{col}_{cat}" for cat in cats]
    names += feature_info["numerical_cols"]
    return names

def print_topk_importances(importances, feature_names, model_name, k=20):
    indices = np.argsort(importances)[::-1][:k]
    print(f"\nTop {k} features for {model_name}:\n")
    for rank, idx in enumerate(indices, 1):
        print(f"{rank:2d}. {feature_names[idx]:<40} {importances[idx]:.4f}")

# â”€â”€â”€ CONFIG â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
REGRESSOR_PICKLE  = "/content/drive/MyDrive/MPCE_MoSPI/Shubham-IIIT/models_regressor/sector_income_randomforestmodel.pkl"
TOP_K             = 20

# â”€â”€â”€ 1. Load everything â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
reg_data     = joblib.load(REGRESSOR_PICKLE)
feature_info = reg_data["feature_info"]
feature_names = get_feature_names(feature_info)

# â”€â”€â”€ 2. Inspect keys to find your four regressors â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
print("Contents of reg_data:", list(reg_data.keys()))
# Also inspect the keys within the 'models' dictionary
print("Contents of reg_data['models']:", list(reg_data['models'].keys()))


# â”€â”€â”€ 3. Define the dict of models you care about â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
# UPDATE these keys to match exactly what you saw above:
# Assuming the keys 1, 2, 3, 4 correspond to the desired models
reg_models = {
    "Lower Rural": reg_data["models"][1],
    "Upper Rural": reg_data["models"][2],
    "Lower Urban": reg_data["models"][3],
    "Upper Urban": reg_data["models"][4],
}

# â”€â”€â”€ 4. Print top-k for each regressor â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
for name, model in reg_models.items():
    if hasattr(model, "feature_importances_"):
        print_topk_importances(model.feature_importances_,
                               feature_names,
                               name,
                               k=TOP_K)
    else:
        print(f"Model {name!r} has no .feature_importances_ attribute")
```

```python
# â”€â”€â”€ 5. Extract top-K features per model into sets â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
top_features = {
    name: set(
        [feature_names[i] for i in np.argsort(model.feature_importances_)[::-1][:TOP_K]]
    )
    for name, model in reg_models.items()
}

# â”€â”€â”€ 6. Compute set differences â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
rural_diff = top_features["Lower Rural"] ^ top_features["Upper Rural"]  # symmetric diff
urban_diff = top_features["Lower Urban"] ^ top_features["Upper Urban"]

rural_exclusive_lower = top_features["Lower Rural"] - top_features["Upper Rural"]
rural_exclusive_upper = top_features["Upper Rural"] - top_features["Lower Rural"]

urban_exclusive_lower = top_features["Lower Urban"] - top_features["Upper Urban"]
urban_exclusive_upper = top_features["Upper Urban"] - top_features["Lower Urban"]

# â”€â”€â”€ 7. Print results â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
print("\nUnique features (Lower vs Upper Rural):")
print(f"  Lower Rural only: {sorted(rural_exclusive_lower)}")
print(f"  Upper Rural only: {sorted(rural_exclusive_upper)}")
print(f"  Symmetric difference: {sorted(rural_diff)}")

print("\nUnique features (Lower vs Upper Urban):")
print(f"  Lower Urban only: {sorted(urban_exclusive_lower)}")
print(f"  Upper Urban only: {sorted(urban_exclusive_upper)}")
print(f"  Symmetric difference: {sorted(urban_diff)}")
```

```python
import pandas as pd
import numpy as np
import joblib

# â”€â”€â”€ HELPERS â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
def get_feature_names(feature_info):
    names = []
    for col in feature_info["categorical_cols"]:
        cats = feature_info["encoders"][col].categories_[0]
        names += [f"{col}_{cat}" for cat in cats]
    names += feature_info["numerical_cols"]
    return names

def topk_features(importances, feature_names, k=20):
    idxs = np.argsort(importances)[::-1][:k]
    return [feature_names[i] for i in idxs]

def print_segment_correlations(df, features, segment_flag, segment_name):
    """
    df:     full training DataFrame
    features: list of feature-column names
    segment_flag: name of the binary column youâ€™ll create
    segment_name: human label for printing
    """
    corrs = []
    for feat in features:
        # Check if feature exists in the DataFrame columns before calculating correlation
        if feat in df.columns:
            corr = df[[feat, segment_flag]].corr().loc[feat, segment_flag]
            corrs.append((feat, corr))
        else:
            print(f"Warning: Feature '{feat}' not found in DataFrame. Skipping correlation calculation.")

    corrs.sort(key=lambda x: abs(x[1]), reverse=True)

    print(f"\nTop correlations vs {segment_name}:\n")
    for feat, corr in corrs:
        print(f"{feat:<40s} {corr:.4f}")


# â”€â”€â”€ CONFIG â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
REGRESSOR_PICKLE   = "/content/drive/MyDrive/MPCE_MoSPI/Shubham-IIIT/models_regressor/sector_income_randomforestmodel.pkl"
TRAIN_WITH_CLASS   = "/content/drive/MyDrive/MPCE_MoSPI/Shubham-IIIT/train_with_class.csv"
TOP_K               = 20

# â”€â”€â”€ 1. Load regressor dict & feature names â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
reg_data       = joblib.load(REGRESSOR_PICKLE)
feature_info   = reg_data["feature_info"]
feature_names  = get_feature_names(feature_info)

# â”€â”€â”€ 2. Identify your four models in the pickle â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
# (use the real keys you find here)
print("Regressor keys:", list(reg_data.keys()))
reg_models = {
    "Lower Rural": reg_data["models"][1],
    "Upper Rural": reg_data["models"][2],
    "Lower Urban": reg_data["models"][3],
    "Upper Urban": reg_data["models"][4],
}

# â”€â”€â”€ 3. Pull top-K features for each regressor â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
top_feats = {
    name: topk_features(m.feature_importances_, feature_names, TOP_K)
    for name, m in reg_models.items()
}

# â”€â”€â”€ 4. Load full training set â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
df = pd.read_csv(TRAIN_WITH_CLASS)

# â”€â”€â”€ 5. For each segment, create a binary flag & compute corrs â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
#    Income_Class mapping: 1=lower_rural, 2=upper_rural, 3=lower_urban, 4=upper_urban
df["is_lower_rural"] = (df["Income_Class"] == 1).astype(int)
df["is_upper_rural"] = (df["Income_Class"] == 2).astype(int)
df["is_lower_urban"] = (df["Income_Class"] == 3).astype(int)
df["is_upper_urban"] = (df["Income_Class"] == 4).astype(int)

for seg, feats in top_feats.items():
    # Filter the features list to only include columns present in the DataFrame
    present_feats = [feat for feat in feats if feat in df.columns]

    # Correctly format segment_flag to match the lowercase column names
    # Use lower() to ensure the segment part is lowercase
    segment_flag = f"is_{seg.lower().replace(' ', '_')}"

    # Check if the constructed segment_flag exists in the DataFrame
    if segment_flag not in df.columns:
        print(f"Error: Segment flag column '{segment_flag}' not found in DataFrame.")
        continue # Skip this iteration if the flag column is missing

    # Also include the segment_flag column in the subset for dropna
    subset_cols = present_feats + [segment_flag]

    # Check if subset_cols is not empty before calling dropna
    if not subset_cols:
        print(f"Warning: No relevant features or segment flag found for segment '{seg}'. Skipping correlation calculation.")
        continue

    print_segment_correlations(
        df=df.dropna(subset=subset_cols),
        features=present_feats, # Pass the filtered list to print_segment_correlations
        segment_flag=segment_flag,
        segment_name=seg.replace("_", " ").title()
    )
```

```python
import matplotlib.pyplot as plt

def plot_segment_correlations(corrs, segment_name):
    """
    corrs: list of (feature, correlation) tuples
    segment_name: title of the segment
    """
    # Unpack
    features, values = zip(*corrs)

    # Plot
    plt.figure(figsize=(10, 6))
    bars = plt.barh(features, values, color='skyblue')
    plt.axvline(0, color='gray', linewidth=1)
    plt.xlabel("Correlation")
    plt.title(f"Top Feature Correlations with {segment_name}")
    plt.gca().invert_yaxis()  # Most correlated on top
    plt.tight_layout()
    plt.show()

for seg, feats in top_feats.items():
    # Filter the features list to only include columns present in the DataFrame
    present_feats = [feat for feat in feats if feat in df.columns]

    # Create correct binary segment flag
    segment_flag = f"is_{seg.lower().replace(' ', '_')}"
    if segment_flag not in df.columns:
        print(f"Error: Segment flag column '{segment_flag}' not found in DataFrame.")
        continue

    # Prepare subset for correlation calculation
    subset_cols = present_feats + [segment_flag]
    if not subset_cols:
        print(f"Warning: No relevant features or segment flag found for segment '{seg}'. Skipping correlation calculation.")
        continue

    # Compute correlations
    corr_list = []
    for feat in present_feats:
        corr = df[[feat, segment_flag]].corr().loc[feat, segment_flag]
        corr_list.append((feat, corr))
    corr_list.sort(key=lambda x: abs(x[1]), reverse=True)

    # Print
    print(f"\nTop correlations vs {seg}:\n")
    for feat, corr in corr_list:
        print(f"{feat:<40s} {corr:.4f}")

    # Plot
    plot_segment_correlations(corr_list, seg)
```

## Correlation Matrix
1. Interpreting the Colors:

* Each cell in the matrix corresponds to the correlation coefficient between the feature in its row and the feature in its column.
* The color scale (colorbar) next to the heatmap indicates the strength and direction of the correlation.
* Generally, colors towards one end of the spectrum (e.g., red) indicate strong positive correlations (as one feature increases, the other tends to increase).

2. Reading the Values: While the heatmap gives a visual overview, the correlation matrix DataFrame you printed below the plot provides the exact correlation coefficient values for each pair of features.

* A value of +1 indicates a perfect positive linear correlation.
* A value of -1 indicates a perfect negative linear correlation.
* A value of 0 indicates no linear correlation.
* Values between -1 and +1 indicate the strength of the linear relationship. The closer the value is to -1 or +1, the stronger the correlation.

3. Diagonal: The diagonal of the matrix (where the row feature is the same as the column feature) will always have a value of 1. This makes sense because a feature is perfectly correlated with itself.

4. Symmetry: The correlation matrix is symmetric. The correlation between Feature A and Feature B is the same as the correlation between Feature B and Feature A. So, the values above the diagonal mirror the values below the diagonal.

## What to Look For:

Strong correlations (close to 1 or -1): This indicates that two features are highly related. In some cases, very high correlation between features could suggest multicollinearity, which might be an issue for certain models, though less so for tree-based models like Random Forest. It also tells you that knowing the value of one feature gives you a good idea of the value of the other.
Weak correlations (close to 0): This suggests the features have little to no linear relationship.

```python
from sklearn.metrics import classification_report, confusion_matrix
import pandas as pd

def evaluate_classification():
    # â”€â”€â”€ Load predicted classes â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    preds = pd.read_csv("/content/drive/MyDrive/MPCE_MoSPI/Shubham-IIIT/final_predictions.csv")

    # â”€â”€â”€ Load true class labels â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    test_df = pd.read_csv("/content/drive/MyDrive/MPCE_MoSPI/Shubham-IIIT/data_with_class/test_with_class.csv")
    test_df = test_df[["HH_ID", "Income_Class"]]

    # â”€â”€â”€ Merge to align predictions and true labels â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    merged = preds.merge(test_df, on="HH_ID", how="inner")

    # Split by Rural and Urban
    rural_mask = merged["Income_Class"].isin([1, 2])
    urban_mask = merged["Income_Class"].isin([3, 4])

    # â”€â”€â”€ Rural Evaluation â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    y_true_rural = merged.loc[rural_mask, "Income_Class"]
    y_pred_rural = merged.loc[rural_mask, "Predicted_Class"]

    print("\n===== Classification Report: RURAL =====")
    print(classification_report(
        y_true_rural, y_pred_rural, labels=[1, 2],
        target_names=["Rural-Lower", "Rural-Upper"], digits=4
    ))
    acc_rural = (y_true_rural == y_pred_rural).mean()
    print(f"Accuracy (Rural): {acc_rural:.4f}")
    print("Confusion Matrix (Rural):")
    print(confusion_matrix(y_true_rural, y_pred_rural, labels=[1, 2]))

    # â”€â”€â”€ Urban Evaluation â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    y_true_urban = merged.loc[urban_mask, "Income_Class"]
    y_pred_urban = merged.loc[urban_mask, "Predicted_Class"]

    print("\n===== Classification Report: URBAN =====")
    print(classification_report(
        y_true_urban, y_pred_urban, labels=[3, 4],
        target_names=["Urban-Lower", "Urban-Upper"], digits=4
    ))
    acc_urban = (y_true_urban == y_pred_urban).mean()
    print(f"Accuracy (Urban): {acc_urban:.4f}")
    print("Confusion Matrix (Urban):")
    print(confusion_matrix(y_true_urban, y_pred_urban, labels=[3, 4]))

    print("================================================")

# Run it
evaluate_classification()
```

## 2. Overall Evaluation Metric (Household-Level)

```python
import pandas as pd
import numpy as np
import joblib
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# â”€â”€â”€ HELPERS â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
def mean_absolute_percentage_error(y_true, y_pred):
    mask = y_true != 0
    return np.mean(np.abs((y_true[mask] - y_pred[mask]) / y_true[mask]) * 100)

# â”€â”€â”€ 1. LOAD PREDICTIONS & GROUND TRUTH â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
preds = pd.read_csv("/content/drive/MyDrive/MPCE_MoSPI/Shubham-IIIT/final_predictions.csv")
test  = pd.read_csv("/content/drive/MyDrive/MPCE_MoSPI/Shubham-IIIT/Testing_data.csv")[["HH_ID","TotalExpense"]]
data  = preds.merge(test, on="HH_ID", how="inner")
data["Error"]        = data["Predicted_TotalExpense"] - data["TotalExpense"]
data["AbsError"]     = np.abs(data["Error"])
data["PctError (%)"] = np.where(
    data["TotalExpense"] != 0,
    data["AbsError"] / data["TotalExpense"] * 100,
    np.nan
)

y_true = data["TotalExpense"].values
y_pred = data["Predicted_TotalExpense"].values

# â”€â”€â”€ 2. COMPUTE METRICS â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
rmse = np.sqrt(mean_squared_error(y_true, y_pred))
mae  = mean_absolute_error(y_true, y_pred)
mape = mean_absolute_percentage_error(y_true, y_pred)
r2   = r2_score(y_true, y_pred)

print("\n===== Final Model Evaluation =====")
print(f"Samples evaluated : {len(data)}")
print(f"RMSE   : {rmse:,.2f}")
print(f"MAE    : {mae:,.2f}")
print(f"MAPE   : {mape:.2f}%")
print(f"RÂ²     : {r2:.4f}")
print("===================================")
```

## 3. Class-wise Eval Metrics(Test Set)

```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# â”€â”€â”€ 1. Load predictions (with Predicted_Class) and true totals â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
preds = pd.read_csv("/content/drive/MyDrive/MPCE_MoSPI/Shubham-IIIT/final_predictions.csv")
test_truth = pd.read_csv(r"/content/drive/MyDrive/MPCE_MoSPI/Shubham-IIIT/Testing_data.csv",
                         usecols=["HH_ID","TotalExpense"]) \
               .rename(columns={"TotalExpense":"Actual_TotalExpense"})

# Merge so each row has HH_ID, Predicted_Class, Predicted_TotalExpense, Actual_TotalExpense
df = preds.merge(test_truth, on="HH_ID", how="inner")

# â”€â”€â”€ 2. Helper for MAPE â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
def mean_absolute_percentage_error(y_true, y_pred):
    mask = y_true != 0
    return np.mean(np.abs((y_true[mask] - y_pred[mask]) / y_true[mask]) * 100)

# â”€â”€â”€ 3. Compute metrics grouped by Predicted_Class â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
rows = []
class_names = {
    1: "Lower Rural",
    2: "Upper Rural",
    3: "Lower Urban",
    4: "Upper Urban"
}

for cls_id, cls_label in class_names.items():
    sub = df[df["Predicted_Class"] == cls_id]
    if sub.empty:
        continue

    y_true = sub["Actual_TotalExpense"].values
    y_pred = sub["Predicted_TotalExpense"].values

    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    mae  = mean_absolute_error(y_true, y_pred)
    mape = mean_absolute_percentage_error(y_true, y_pred)
    r2   = r2_score(y_true, y_pred)

    rows.append({
        "Class_ID": cls_id,
        "Class_Name": cls_label,
        "N_Samples": len(sub),
        "RMSE": rmse,
        "MAE": mae,
        "MAPE (%)": mape,
        "RÂ²": r2
    })

results_df = pd.DataFrame(rows,
                          columns=["Class_ID","Class_Name","N_Samples","RMSE","MAE","MAPE (%)","RÂ²"])

# â”€â”€â”€ 4. Display the table â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
print("\nClass-wise Metrics (by Predicted_Class):\n")
print(results_df.to_string(index=False))
```

# The "Residuals vs. Actual TotalExpense" plot:
Generated is a valuable tool for evaluating the performance of your regression model. Here's what it signifies:

* Residuals: A residual is the difference between the actual (true) value of a data point and the value predicted by your model. In your plot, this is represented by the "Prediction Error (Pred - True)" on the y-axis.

* Positive residuals mean the model underpredicted the actual value.
* Negative residuals mean the model overpredicted the actual value.
* Residuals close to zero mean the model's prediction was close to the actual value.

# Key Observations
1. Residuals vs. Actual TotalExpense
ðŸ“Š Plot Summary:
* X-axis: Actual TotalExpense (ground truth)

* Y-axis: Prediction Error (i.e., y_pred - y_true)

* The dashed horizontal line represents zero error.

* Points below the line = underestimation; above = overestimation.

ðŸ” Observations:
* Heteroscedasticity: The residuals (errors) spread more as TotalExpense increases. This fan-shaped pattern means the model is less accurate for higher values.

* Bias: Most residuals are negative at higher expense values â†’ the model consistently underpredicts for high-expense households.

* Density: A large number of data points are clustered around lower expense values, indicating data imbalance (more low-income samples).

ðŸ“‰ Implications:
The model struggles with generalizing on high-expense households.

You may need:

* Log-transforming TotalExpense to stabilize variance.

* Using quantile regression or heteroscedastic models.

* Balancing the dataset or stratifying based on expense class.

ðŸ“‰ Implications:

Consider:

* Trimming or capping extreme predictions.

* Investigating if misclassified classes (e.g., 3 vs. 4) have structural issues.

```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import joblib

def main():
    # â”€â”€â”€ CONFIG â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    PRED_CSV = "/content/drive/MyDrive/MPCE_MoSPI/Shubham-IIIT/final_predictions.csv"         # output of Step 3
    TEST_CSV = "/content/drive/MyDrive/MPCE_MoSPI/Shubham-IIIT/Testing_data.csv"  # original with TotalExpense

    # â”€â”€â”€ 1. Load data â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    preds = pd.read_csv(PRED_CSV)
    test  = pd.read_csv(TEST_CSV)[["HH_ID", "TotalExpense"]]

    # â”€â”€â”€ 2. Merge on HH_ID â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    merged = preds.merge(test, on="HH_ID", how="inner")
    if len(merged) != len(preds):
        print("Warning: some HH_IDs in predictions not matched in test set.")

    # â”€â”€â”€ 3. Extract arrays â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    y_true = merged["TotalExpense"].values
    y_pred = merged["Predicted_TotalExpense"].values

    # â”€â”€â”€ 4. Compute metrics â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    mae  = mean_absolute_error(y_true, y_pred)
    mape = mean_absolute_percentage_error(y_true, y_pred)
    r2   = r2_score(y_true, y_pred)

    # â”€â”€â”€ 5. Print results â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    print("\n===== Final Model Evaluation =====")
    print(f"Samples evaluated : {len(merged)}")
    print(f"RMSE   : {rmse:,.2f}")
    print(f"MAE    : {mae:,.2f}")
    print(f"MAPE   : {mape:.2f}%")
    print(f"RÂ²     : {r2:.4f}")
    print("===================================")

    # Optional: show a few rows
    print("\nPredicted vs Actual:")
    print(merged[["HH_ID", "Predicted_TotalExpense", "TotalExpense"]].head(50))

if __name__ == "__main__":
    main()
```

```python
import pandas as pd

# â”€â”€â”€ 1. Load the merged predictions+weights+actual file â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
df = pd.read_csv("/content/drive/MyDrive/MPCE_MoSPI/Shubham-IIIT/predictions_with_weights_and_actual.csv")

# â”€â”€â”€ 2. Identify the weight column â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
weight_cols = [c for c in df.columns if "weight" in c.lower()]
if not weight_cols:
    raise ValueError("No weight column found. Please ensure your merged CSV has a weight column.")
weight_col = weight_cols[0]
print(f"Using weight column: {weight_col}")

# â”€â”€â”€ 3. Bring in household size from the test file â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
test_df = pd.read_csv("/content/drive/MyDrive/MPCE_MoSPI/Shubham-IIIT/Testing_data.csv")
df = df.merge(
    test_df[["HH_ID", "HH Size (For FDQ)"]],
    on="HH_ID", how="left"
)
if df["HH Size (For FDQ)"].isna().any():
    print(f"Warning: {df['HH Size (For FDQ)'].isna().sum()} HHs missing size.")

# â”€â”€â”€ 4. Compute weighted sums â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
# For actual TotalExpense
df["WtdTotalExpense_actual"]  = df["Actual_TotalExpense"] * df[weight_col]
# For predicted TotalExpense
df["WtdTotalExpense_predict"] = df["Predicted_TotalExpense"] * df[weight_col]
# Weighted HHâ€size
df["WtdHH_in_FDQ"]            = df["HH Size (For FDQ)"]   * df[weight_col]

# â”€â”€â”€ 5. Group by State and compute MPCE via sumâ€ratio â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
# Build a small helper
def mpce_by(col):
    sums_exp = df.groupby("State")[col].sum()
    sums_size = df.groupby("State")["WtdHH_in_FDQ"].sum()
    return (sums_exp / sums_size).reset_index(name=col.replace("WtdTotalExpense_", "MPCE_"))

pred_mpce = mpce_by("WtdTotalExpense_predict").rename(columns={"MPCE_WtdTotalExpense_predict": "Predicted_MPCE"})
act_mpce  = mpce_by("WtdTotalExpense_actual").rename(columns={"MPCE_WtdTotalExpense_actual":  "Actual_MPCE"})

# â”€â”€â”€ 6. Combine and save â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
state_mpce = pred_mpce.merge(act_mpce, on="State")
state_mpce.to_csv(
    "/content/drive/MyDrive/MPCE_MoSPI/Shubham-IIIT/statewise_MPCE.csv",
    index=False
)
print("âœ” Saved statewise MPCE to statewise_MPCE.csv")

# â”€â”€â”€ 7. Display â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
print("\nState-wise MPCE:")
print(state_mpce.to_string(index=False))
```

```python
import pandas as pd
import numpy as np
from IPython.display import display

# â”€â”€â”€ 1. Load prediction + actuals â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
df = pd.read_csv("/content/drive/MyDrive/MPCE_MoSPI/Shubham-IIIT/predictions_with_weights_and_actual.csv")

# â”€â”€â”€ 2. Identify weight column â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
weight_cols = [c for c in df.columns if "weight" in c.lower()]
if not weight_cols:
    raise ValueError("No weight column found.")
weight_col = weight_cols[0]
print(f"Using weight column: {weight_col}")

# â”€â”€â”€ 3. Merge with HH Size (For FDQ) â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
test_df = pd.read_csv("/content/drive/MyDrive/MPCE_MoSPI/Shubham-IIIT/Testing_data.csv")
df = df.merge(test_df[["HH_ID", "HH Size (For FDQ)"]], on="HH_ID", how="left")
if df["HH Size (For FDQ)"].isna().any():
    print(f"âš  Warning: {df['HH Size (For FDQ)'].isna().sum()} HHs missing 'HH Size (For FDQ)'")

# â”€â”€â”€ 4. Compute weighted values â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
df["WtdTotalExpense_predict"] = df["Predicted_TotalExpense"] * df[weight_col]
df["WtdTotalExpense_actual"]  = df["Actual_TotalExpense"]    * df[weight_col]
df["WtdHHSize_FDQ"]           = df["HH Size (For FDQ)"]      * df[weight_col]

# â”€â”€â”€ 5. Group by State and Sector â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
grouped = df.groupby(["State", "Sector"])
pred_mpce = grouped["WtdTotalExpense_predict"].sum() / grouped["WtdHHSize_FDQ"].sum()
act_mpce  = grouped["WtdTotalExpense_actual"].sum()  / grouped["WtdHHSize_FDQ"].sum()

state_sector_mpce = pd.DataFrame({
    "Predicted_MPCE": pred_mpce,
    "Actual_MPCE": act_mpce
}).reset_index()

# â”€â”€â”€ 6. Map State and Sector Names â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
state_labels = {
    1: "Jammu & Kashmir", 2: "Himachal Pradesh", 3: "Punjab", 4: "Chandigarh (U.T.)",
    5: "Uttarakhand", 6: "Haryana", 7: "Delhi", 8: "Rajasthan", 9: "Uttar Pradesh",
    10: "Bihar", 11: "Sikkim", 12: "Arunachal Pradesh", 13: "Nagaland", 14: "Manipur",
    15: "Mizoram", 16: "Tripura", 17: "Meghalaya", 18: "Assam", 19: "West Bengal",
    20: "Jharkhand", 21: "Odisha", 22: "Chhattisgarh", 23: "Madhya Pradesh", 24: "Gujarat",
    25: "Dadra & Nagar Haveli and Daman & Diu", 26: "Telangana", 27: "Maharashtra",
    28: "Andhra Pradesh", 29: "Karnataka", 30: "Goa", 31: "Lakshadweep (U.T.)",
    32: "Kerala", 33: "Tamil Nadu", 34: "Puducherry (U.T.)",
    35: "Andaman & Nicobar Islands (U.T.)", 36: "Ladakh (U.T.)"
}
sector_labels = {1: "Rural", 2: "Urban"}

state_sector_mpce["StateName"] = state_sector_mpce["State"].map(state_labels)
state_sector_mpce["SectorName"] = state_sector_mpce["Sector"].map(sector_labels)

# â”€â”€â”€ 7. Compute % Error â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
state_sector_mpce["%Error"] = 100 * np.abs(
    state_sector_mpce["Predicted_MPCE"] - state_sector_mpce["Actual_MPCE"]
) / state_sector_mpce["Actual_MPCE"]

# â”€â”€â”€ 8. Save to CSV â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
state_sector_mpce.to_csv(
    "/content/drive/MyDrive/MPCE_MoSPI/Shubham-IIIT/state_sector_MPCE.csv", index=False
)
print("âœ” Saved state+sector MPCE with %Error to CSV\n")

# â”€â”€â”€ 9. Display Styled Table â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
cols_to_show = ["State", "StateName", "Sector", "SectorName", "Predicted_MPCE", "Actual_MPCE", "%Error"]
styled = state_sector_mpce[cols_to_show].style \
    .format({
        "Predicted_MPCE": "{:.2f}",
        "Actual_MPCE": "{:.2f}",
        "%Error": "{:.2f}%"
    }) \
    .set_table_styles([
        {"selector": "th", "props": [("font-size", "14pt"), ("font-weight", "bold")]}
    ])

display(styled)
```

```python
import pandas as pd

# â”€â”€â”€ 1. Load predictions file â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
df = pd.read_csv("/content/drive/MyDrive/MPCE_MoSPI/Shubham-IIIT/predictions_with_weights_and_actual.csv")

# â”€â”€â”€ 2. Identify the weight column â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
weight_cols = [c for c in df.columns if "weight" in c.lower()]
if not weight_cols:
    raise ValueError("No weight column found. Please ensure your merged CSV has a weight column.")
weight_col = weight_cols[0]
print(f"Using weight column: {weight_col}")

# â”€â”€â”€ 3. Bring in household size + NSS Region â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
test_df = pd.read_csv("/content/drive/MyDrive/MPCE_MoSPI/Shubham-IIIT/Testing_data.csv")

df = df.merge(
    test_df[["HH_ID", "HH Size (For FDQ)", "NSS-Region"]],
    on="HH_ID", how="left"
)

if df["HH Size (For FDQ)"].isna().any():
    print(f"Warning: {df['HH Size (For FDQ)'].isna().sum()} HHs missing HH size.")
if df["NSS-Region"].isna().any():
    print(f"Warning: {df['NSS-Region'].isna().sum()} HHs missing NSS region.")

# â”€â”€â”€ 4. Compute weighted sums â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
df["WtdTotalExpense_actual"]  = df["Actual_TotalExpense"] * df[weight_col]
df["WtdTotalExpense_predict"] = df["Predicted_TotalExpense"] * df[weight_col]
df["WtdHH_in_FDQ"]            = df["HH Size (For FDQ)"]   * df[weight_col]

# â”€â”€â”€ 5. Group by NSS Region to compute MPCE â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
def mpce_by_region(col):
    sums_exp = df.groupby("NSS-Region")[col].sum()
    sums_size = df.groupby("NSS-Region")["WtdHH_in_FDQ"].sum()
    return (sums_exp / sums_size).reset_index(name=col.replace("WtdTotalExpense_", "MPCE_"))

region_pred_mpce = mpce_by_region("WtdTotalExpense_predict").rename(
    columns={"MPCE_WtdTotalExpense_predict": "Predicted_MPCE"}
)
region_act_mpce = mpce_by_region("WtdTotalExpense_actual").rename(
    columns={"MPCE_WtdTotalExpense_actual": "Actual_MPCE"}
)

# â”€â”€â”€ 6. Merge and Save â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
region_mpce = region_pred_mpce.merge(region_act_mpce, on="NSS-Region")
region_mpce.to_csv(
    "/content/drive/MyDrive/MPCE_MoSPI/Shubham-IIIT/regionwise_MPCE.csv",
    index=False
)
print("âœ” Saved region-wise MPCE to regionwise_MPCE.csv")

# â”€â”€â”€ 7. Display â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
print("\nRegion-wise MPCE:")
print(region_mpce.to_string(index=False))
```

# State + District Wise MPCE-Predictive Modelling:

```python
import pandas as pd

# â”€â”€â”€ 1. Load merged predictions + actuals â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
df = pd.read_csv("/content/drive/MyDrive/MPCE_MoSPI/Shubham-IIIT/predictions_with_weights_and_actual.csv")
print(f"Loaded merged predictions: {len(df)} rows")

# â”€â”€â”€ 2. Load test file to get HH Size (For FDQ) â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
test_df = pd.read_csv("/content/drive/MyDrive/MPCE_MoSPI/Shubham-IIIT/Testing_data.csv",
                      usecols=["HH_ID", "HH Size (For FDQ)", "District"])
df = df.merge(test_df, on="HH_ID", how="left")
missing = df["HH Size (For FDQ)"].isna().sum()
if missing:
    print(f"Warning: {missing} households missing HH Size (For FDQ)")

# â”€â”€â”€ 3. Identify the weight column â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
weight_cols = [c for c in df.columns if "weight" in c.lower()]
if not weight_cols:
    raise ValueError("No weight column found in your merged CSV.")
weight_col = weight_cols[0]
print(f"Using weight column: {weight_col}")

# â”€â”€â”€ 4. Create weightedâ€expense and weightedâ€size columns â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
df["WtdPredict"]    = df["Predicted_TotalExpense"] * df[weight_col]
df["WtdActual"]     = df["Actual_TotalExpense"]    * df[weight_col]
df["WtdHHSize_FDQ"] = df["HH Size (For FDQ)"]      * df[weight_col]

# â”€â”€â”€ 5. Group by State+District and compute MPCE via sumâ€ratio â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
grp = df.groupby(["State","District"])
pred_mpce = grp["WtdPredict"].sum()   / grp["WtdHHSize_FDQ"].sum()
act_mpce  = grp["WtdActual"].sum()    / grp["WtdHHSize_FDQ"].sum()

state_district = (
    pd.DataFrame({
        "Predicted_MPCE": pred_mpce,
        "Actual_MPCE":    act_mpce
    })
    .reset_index()
)

# â”€â”€â”€ 6. Map state codes to names (optional) â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
state_options = {
    1: "Jammu & Kashmir", 2: "Himachal Pradesh", 3: "Punjab", 4: "Chandigarh (U.T.)",
    5: "Uttarakhand", 6: "Haryana", 7: "Delhi", 8: "Rajasthan", 9: "Uttar Pradesh",
    10: "Bihar", 11: "Sikkim", 12: "Arunachal Pradesh", 13: "Nagaland", 14: "Manipur",
    15: "Mizoram", 16: "Tripura", 17: "Meghalaya", 18: "Assam", 19: "West Bengal",
    20: "Jharkhand", 21: "Odisha", 22: "Chhattisgarh", 23: "Madhya Pradesh", 24: "Gujarat",
    25: "Dadra & Nagar Haveli and Daman & Diu", 26: "Telangana", 27: "Maharashtra",
    28: "Andhra Pradesh", 29: "Karnataka", 30: "Goa", 31: "Lakshadweep (U.T.)",
    32: "Kerala", 33: "Tamil Nadu", 34: "Puducherry (U.T.)",
    35: "Andaman & Nicobar Islands (U.T.)", 36: "Ladakh (U.T.)"
}
state_district["StateName"] = state_district["State"].map(state_options)

# â”€â”€â”€ 7. Save and display â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
out_cols = ["State","StateName","District","Predicted_MPCE","Actual_MPCE"]
state_district.to_csv(
    "/content/drive/MyDrive/MPCE_MoSPI/Shubham-IIIT/state_district_MPCE.csv",
    index=False
)
print(f"âœ” Saved state+district MPCE (sumâ€‘ratio) to state_district_MPCE.csv\n")

print("Sample state+district MPCE (Predicted vs Actual):")
print(state_district[out_cols].head(10).to_string(index=False))
```

```python
import pandas as pd
import numpy as np
import joblib
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# â”€â”€â”€ CONFIG â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
RAW_TEST_CSV           = "/content/drive/MyDrive/MPCE_MoSPI/HCES2023-24/merged_final_data.csv"
UNLABELED_TEST_CSV     = "test_raw.csv"
REGRESSOR_PICKLE       = "/content/drive/MyDrive/MPCE_MoSPI/Shubham-IIIT/models_Regressor/sector_income_randomforestmodel.pkl"
CLASSIFIER_PICKLE      = "/content/drive/MyDrive/MPCE_MoSPI/Shubham-IIIT/models_Clf/sector_income_classifiers_tuned.pkl"
OUTPUT_PRED_CSV        = "/content/drive/MyDrive/MPCE_MoSPI/Shubham-IIIT/final_predictions_2023_24.csv"

# â”€â”€â”€ 1. Prepare unlabeled test set â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
raw = pd.read_csv(RAW_TEST_CSV)
# save true labels for later
true_labels = raw[["HH_ID", "TotalExpense"]].copy()
# drop target for inference
test_df = raw.drop(columns=["TotalExpense"])
test_df.to_csv(UNLABELED_TEST_CSV, index=False)
print(f"âœ” Saved raw test set (no target) â†’ {UNLABELED_TEST_CSV}")

# â”€â”€â”€ 2. Load unlabeled test set â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
test_df = pd.read_csv(UNLABELED_TEST_CSV)
print(f"Loaded raw test set: {len(test_df)} rows")

# â”€â”€â”€ 3. Load regressors + preprocessing info â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
reg_data   = joblib.load(REGRESSOR_PICKLE)
reg_models = reg_data["models"]        # {1: reg1, â€¦, 4: reg4}
feat_info  = reg_data["feature_info"]  # encoders, scaler, etc.

# â”€â”€â”€ 4. Load classifiers â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
clf_data  = joblib.load(CLASSIFIER_PICKLE)
clf_rural = clf_data["clf_rural"]
clf_urban = clf_data["clf_urban"]

# â”€â”€â”€ 5. Batch feature preparation â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
def prepare_features_batch(data, feature_info):
    cats = feature_info["categorical_cols"]
    nums = feature_info["numerical_cols"]
    encs = feature_info["encoders"]
    scaler = feature_info["scaler"]

    parts = []
    for col in cats:
        if col in data:
            parts.append(encs[col].transform(data[[col]]))
    if nums:
        parts.append(scaler.transform(data[nums]))
    return np.hstack(parts)

X_all = prepare_features_batch(test_df, feat_info)

# â”€â”€â”€ 6. Predict income class â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
predicted_classes = np.zeros(len(test_df), dtype=int)
mask_r = test_df["Sector"] == 1
mask_u = test_df["Sector"] == 2

if mask_r.any():
    r_preds = clf_rural.predict(X_all[mask_r])
    predicted_classes[mask_r] = np.where(r_preds==1, 2, 1)

if mask_u.any():
    u_preds = clf_urban.predict(X_all[mask_u])
    predicted_classes[mask_u] = np.where(u_preds==1, 4, 3)

# â”€â”€â”€ 7. Predict total expense per class â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
pred_exp = np.zeros(len(test_df))
for cls_id, model in reg_models.items():
    idx = predicted_classes == cls_id
    if idx.any():
        pred_exp[idx] = model.predict(X_all[idx])

# â”€â”€â”€ 8. Build prediction DataFrame & drop duplicate HH_IDs â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
output_df = pd.DataFrame({
    "HH_ID": test_df["HH_ID"],
    "Predicted_Class": predicted_classes,
    "Predicted_TotalExpense": pred_exp
})
before = len(output_df)
# â”€â”€â”€ 8. Drop duplicate predictions â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
output_df = output_df.drop_duplicates(subset="HH_ID", keep="first")

# â”€â”€â”€ 9. Deduplicate true labels â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
true_labels = true_labels.drop_duplicates(subset="HH_ID", keep="first")

# â”€â”€â”€ 10. Merge for evaluation â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
merged = output_df.merge(true_labels, on="HH_ID", how="inner")
missing = len(output_df) - len(merged)
print(f"Warning: {missing} HH_IDs in predictions not matched in ground truth")

# â”€â”€â”€ 11. Compute metrics â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
y_true = merged["TotalExpense"].values
y_pred = merged["Predicted_TotalExpense"].values

rmse = np.sqrt(mean_squared_error(y_true, y_pred))
mae  = mean_absolute_error(y_true, y_pred)
mape = np.mean(np.abs((y_true - y_pred) / y_true) * 100)
r2   = r2_score(y_true, y_pred)

print("\n===== Final Model Evaluation =====")
print(f"Samples evaluated : {len(merged)}")
print(f"RMSE   : {rmse:,.2f}")
print(f"MAE    : {mae:,.2f}")
print(f"MAPE   : {mape:.2f}%")
print(f"RÂ²     : {r2:.4f}")
print("===============================")

# â”€â”€â”€ 10. Optional: scatter plot actual vs predicted â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
plt.figure(figsize=(6,6))
plt.scatter(y_true, y_pred, alpha=0.3, s=5)
lims = [0, 400000]
plt.plot(lims, lims, 'r--')
plt.xlabel("Actual TotalExpense")
plt.ylabel("Predicted TotalExpense")
plt.title("Actual vs Predicted (2023â€‘24)")
plt.xlim(lims); plt.ylim(lims)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error

# â”€â”€â”€ 1. Load stateâ€‘wise MPCE â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
df = pd.read_csv("/content/drive/MyDrive/MPCE_MoSPI/Shubham-IIIT/results/statewise_MPCE_2023_24.csv")

# â”€â”€â”€ 2. Extract arrays â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
y_true = df["Actual_MPCE"].values
y_pred = df["Predicted_MPCE"].values

# â”€â”€â”€ 3. Compute RMSE â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
rmse = np.sqrt(mean_squared_error(y_true, y_pred))

# â”€â”€â”€ 4. Compute MAPE â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
# mask zeros to avoid division by zero
mask = y_true != 0
mape = np.mean(np.abs((y_true[mask] - y_pred[mask]) / y_true[mask])) * 100

# â”€â”€â”€ 5. Print results â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
print("Stateâ€‘wise MPCE Evaluation Metrics")
print("----------------------------------")
print(f"Number of states evaluated : {len(df)}")
print(f"RMSE  : {rmse:,.2f}")
print(f"MAPE  : {mape:.2f}%")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# â”€â”€â”€ 1. Load the State+Sector MPCE table â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
state_sector = pd.read_csv("/content/drive/MyDrive/MPCE_MoSPI/Shubham-IIIT/results/state_sector_MPCE_2023_24.csv")

# â”€â”€â”€ 2. Split out Rural vs Urban â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
rural = state_sector[state_sector["Sector"] == 1].sort_values("State")
urban = state_sector[state_sector["Sector"] == 2].sort_values("State")

# Map state codes to names if not already present
state_labels = {
    1: "Jammu & Kashmir", 2: "Himachal Pradesh", 3: "Punjab", 4: "Chandigarh",
    5: "Uttarakhand", 6: "Haryana", 7: "Delhi", 8: "Rajasthan", 9: "Uttar Pradesh",
    10:"Bihar", 11:"Sikkim", 12:"Arunachal Pradesh", 13:"Nagaland", 14:"Manipur",
    15:"Mizoram",16:"Tripura",17:"Meghalaya",18:"Assam",19:"West Bengal",
    20:"Jharkhand",21:"Odisha",22:"Chhattisgarh",23:"Madhya Pradesh",24:"Gujarat",
    25:"Dadra & Nagar Haveli & Daman & Diu",26:"Telangana",27:"Maharashtra",
    28:"Andhra Pradesh",29:"Karnataka",30:"Goa",31:"Lakshadweep",32:"Kerala",
    33:"Tamil Nadu",34:"Puducherry",35:"Andaman & Nicobar",36:"Ladakh"
}
rural["StateName"] = rural["State"].map(state_labels)
urban["StateName"] = urban["State"].map(state_labels)

# â”€â”€â”€ 3. Rural Plot â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
plt.figure(figsize=(14,6))
x_r = np.arange(len(rural))
width = 0.35

plt.bar(x_r - width/2, rural["Predicted_MPCE"], width, label="Predicted")
plt.bar(x_r + width/2, rural["Actual_MPCE"],    width, label="Actual")

plt.xticks(x_r, rural["StateName"], rotation=90, fontsize=8)
plt.title("Rural MPCE by State")
plt.xlabel("State")
plt.ylabel("MPCE")
plt.grid(axis="y", linestyle="--", alpha=0.5)
plt.legend()
plt.tight_layout()
plt.show()

# â”€â”€â”€ 4. Urban Plot â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
plt.figure(figsize=(14,6))
x_u = np.arange(len(urban))

plt.bar(x_u - width/2, urban["Predicted_MPCE"], width, label="Predicted", color="#ff7f0e")
plt.bar(x_u + width/2, urban["Actual_MPCE"],    width, label="Actual",    color="#1f77b4")

plt.xticks(x_u, urban["StateName"], rotation=90, fontsize=8)
plt.title("Urban MPCE by State")
plt.xlabel("State")
plt.ylabel("MPCE")
plt.grid(axis="y", linestyle="--", alpha=0.5)
plt.legend()
plt.tight_layout()
plt.show()
```
